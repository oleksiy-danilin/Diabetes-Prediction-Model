<span style="font-size: 16px;">In this book, we check correlations of the diabetes/prediabetes prevalence ('diabetes_binary' feature) with various categorical variables, as well as with binary (yes/no) variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

### **Correlation with Categorical Variables**

```python
## Initial dataset
df = pd.read_csv('diabetes_binary.csv')
```

```python
df.head()
```

```python
# Specify the columns you want to include in the correlation heatmap
subset = df[['Diabetes_binary', 'BMI', 'GenHlth', 'MentHlth', 'PhysHlth', 'Sex', 'Age', 'Education', 'Income']]  # Replace with your actual column names

# Compute the correlation matrix for the selected variables
corr = subset.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))  # Adjust the size to improve readability
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

plt.title('Correlation Heatmap: Diabetes Prevalence with Categorical Variables')
plt.show()
```

### **Correlation with Binary Variables** 

```python
df = pd.read_csv('diabetes_ONLY_binary.csv')
```

```python
df.head()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix for the numeric columns
corr = numeric_df.corr()

# Plot the heatmap with a larger size and better readability
plt.figure(figsize=(16, 14))  # Increase the figure size to make it larger
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, annot_kws={"size": 10}, linewidths=0.5)

# Add title and adjust font size
plt.title('Correlation Heatmap: Diabetes Prevalence with Binary Variables', fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```
